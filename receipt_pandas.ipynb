```python
import json
import pandas as pd
import numpy as np
import datetime
import time
from tqdm import tqdm, tqdm_notebook
```

```python
# 字典轉list
def user_dict_list():
    user_dict_list=[]
    with open('data/jieba_data/user_receipt.txt','r',encoding='utf-8') as f:
        for line in f:
            user_dict_list.append(line.strip('\n'))
    return user_dict_list
print(user_dict_list())
```

```python
# 讀取7G多檔案大小。更改運算邏輯 (iterrows 改 str.contains) 從8小時降為3分鐘

# 因df資料較大，使用chunk進行區塊處理，中間將不包含指定list內容去掉
tStart = time.time()
df = pd.read_csv('data/invoice_qmonster.csv',header=None,encoding='utf8', chunksize=10000)
for chunk in df:
    chunk.columns=['inv_id','item_no','product_name','unit_price','quantity','amonut','inv_time','login_time','company_name','seller_address','unknow','gender','age']
    chunk = chunk[chunk['product_name'].str.contains('|'.join(user_dict_list()), na=False)]  # na : default NaN, fill value for missing values.
    chunk.to_csv('testtest.csv', mode='a',index=0, header=0,encoding='utf-8-sig')
    print(chunk['inv_time'])
tEnd = time.time()
print('花費時間',tEnd-tStart)
```
