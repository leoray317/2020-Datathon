```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('store_merge_detail.csv')
```

```python
df
```

```python
df.columns=['inv_id','item_no','product_name','unit_price','quantity','amonut','inv_time','login_time','company_name','seller_address','unknow','gender','age','store_name','ad1','ad2']

# 將時間欄位以空白拆成新的兩欄位
df[['inv_date','inv_date_time']] = df.inv_time.str.split(" ",expand=True,)
```

```python
# 分開台北與以外的資料
df_o = df[df['ad1']!='台北市']
df_t = df[df['ad1']=='台北市']
```

```python
df_o_list = df_o['ad1'].drop_duplicates().tolist()
print(df_o_list)
df_t_list = df_t['ad2'].drop_duplicates().tolist()
print(df_t_list)
```

```python
#台北以外區域地點
df_o1 = df_o[['inv_date','ad1','quantity']]
df_o1_g = df_o1.groupby(['inv_date','ad1']).sum()
print(df_o1_g )
```

```python
#台北區域地點
df_t1 = df_t[['inv_date','ad2','quantity']]
df_t1_g = df_t1.groupby(['inv_date','ad2']).sum()
print(df_t1_g)
```

```python
#台北以外區域地點
date_area_o1=[]
for i,j in enumerate(df_o1_g.values):
    date_area_o1.append([df_o1_g.index[i][0],df_o1_g.index[i][1],j[0]])
d_area_o1 = pd.DataFrame(date_area_o1,columns=['inv_date','ad1','quantity'])
print(d_area_o1)
```

```python
#台北區域地點
date_area_t1=[]
for i,j in enumerate(df_t1_g.values):
    date_area_t1.append([df_t1_g.index[i][0],df_t1_g.index[i][1],j[0]])
d_area_t1=pd.DataFrame(date_area_t1,columns=['inv_date','ad2','quantity'])
print(d_area_t1)
```

```python
#台北以外區域地點
df_dict_o1 = dict()
for i in df_o_list:
    df_dict_o1[i] = d_area_o1[d_area_o1['ad1'] == i]
df_dict_o1
```

```python
#台北區域地點
df_dict_t1 = dict()
for i in df_t_list:
    df_dict_t1[i] = d_area_t1[d_area_t1['ad2'] == i]
    print(df_dict_t1[i])
df_dict_t1
```

```python
t = d_area_t1['ad2'].drop_duplicates()
o = d_area_o1['ad1'].drop_duplicates()
t_area=[]
for i in t:
    t_area.append(i)
o_area= []
for i in o:
    o_area.append(i)

t_month=[]
for i in d_area_t1['inv_date'].values:
    t_month.append(i.split('-')[1])
o_month=[]
for i in d_area_o1['inv_date'].values:
    o_month.append(i.split('-')[1])

d_area_t1['ad'] = d_area_t1['ad2']
d_area_t1['month'] = t_month
d_area_o1['ad']=d_area_o1['ad1']
d_area_o1['month'] = o_month

d_area = pd.concat([d_area_t1,d_area_o1])
area = d_area.groupby(['month','ad']).sum()

area_new=[]
for i,j in enumerate(area.values):
    area_new.append([area.index[i][0],area.index[i][1],j[0]])
area_df = pd.DataFrame(area_new,columns = ['time','area','num'])
area_df
```

```python
area = []
for i in area_df.values:
    if i[0] != '07' and i[0] != '08':
        area.append([str(i[1]),int(i[0]),int(i[2])])
area_df = pd.DataFrame(area,columns = ['鄉鎮市區','月份','num'])
```

```python
area_df
```

```python
df = pd.read_csv('rain_garbage_final.csv')
print(df)
gra=[]
for i in df.values:
    gra.append([str(i[0]),int(i[1]),int(i[2]),int(i[3]),i[4]])
gar_df = pd.DataFrame(gra,columns = ['鄉鎮市區','月份','over30C','rain','garbage'])
```

```python
df_total = pd.merge(area_df,gar_df,on=['鄉鎮市區','月份'])
```

```python
a = df_total[df_total['鄉鎮市區'] == '中山區'].sort_values(by= '月份')

print(type(a))
```

```python
import statsmodels.api as sm
#df_csv = pd.read_csv("data/garbage_final.csv", sep=",", encoding="")

#f = pd.DataFrame(a,columns=['num','over30C', 'rain', 'garbage'])

#print(type(f))

x = a[['num','over30C', 'rain']]
y = a[['garbage']]

x = sm.add_constant(x)

model = sm.OLS(y,x).fit()
predictions = model.predict(x)

print(model.summary())
```
